```python
import matplotlib.pyplot as plt
import numpy as np

# Best performance data for RNN, LSTM, CNN (manually extracted from results)
models = ['RNN', 'LSTM', 'CNN']
accuracy = [0.83748, 0.8408, 0.83728]  # Best accuracy from results
precision_0 = [0.84, 0.90, 0.84]  # Precision for class 0
recall_0 = [0.83, 0.76, 0.83]  # Recall for class 0
f1_score_0 = [0.84, 0.83, 0.84]  # F1-score for class 0
precision_1 = [0.83, 0.80, 0.83]  # Precision for class 1
recall_1 = [0.84, 0.92, 0.85]  # Recall for class 1
f1_score_1 = [0.84, 0.85, 0.84]  # F1-score for class 1

# Plot accuracy
plt.figure(figsize=(8, 6))
plt.bar(models, accuracy, color=['blue', 'green', 'orange'])
plt.title('Model Accuracy Comparison')
plt.ylabel('Accuracy')
plt.ylim(0.75, 0.9)
plt.show()

# Plot precision, recall, and F1-score for Class 0
x = np.arange(len(models))
width = 0.2

plt.figure(figsize=(10, 6))
plt.bar(x - width, precision_0, width, label='Precision (0)', color='blue')
plt.bar(x, recall_0, width, label='Recall (0)', color='green')
plt.bar(x + width, f1_score_0, width, label='F1-Score (0)', color='orange')
plt.title('Class 0 Metrics Comparison')
plt.ylabel('Score')
plt.xticks(x, models)
plt.legend()
plt.ylim(0.7, 1.0)
plt.show()

# Plot precision, recall, and F1-score for Class 1
plt.figure(figsize=(10, 6))
plt.bar(x - width, precision_1, width, label='Precision (1)', color='blue')
plt.bar(x, recall_1, width, label='Recall (1)', color='green')
plt.bar(x + width, f1_score_1, width, label='F1-Score (1)', color='orange')
plt.title('Class 1 Metrics Comparison')
plt.ylabel('Score')
plt.xticks(x, models)
plt.legend()
plt.ylim(0.7, 1.0)
plt.show()
```
